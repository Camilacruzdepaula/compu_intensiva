## ¿Qué es NLP?

NLP hace referencia a Natural Language Processing, en español Procesamiento de Lenguaje Natural, y es la rama de la ciencia de la computación que permite construir algoritmos que procesan, analizan e incluso producen el lenguaje tal y como lo hace una persona. Esta herramienta computacional es una combinación entre estadística, aprendizade de máquina, y aprendizaje profundo.

Los aplicaciones más comunes de NLP son:

* Reconocimiento de voz (Convertir un audio de voz a texto)
* Etiquetado gramatical 
* Análisis de sentimiento
* Generación del Lenguaje Natural

### Muestra de contenidos latex

También es importante entender la matemática detrás de los algoritmos para NLP. Como los documentos de texto se suelen representar con vectores, es fundamental tener claros los conceptos básicos de álgebral lineal. A continuación, se presentan algunas fórmulas de utilidad:

* Producto Punto:

$$\vec{p}=(x_{1}, y_{1},z_{1})$$
  $$\vec{q}=(x_{2}, y_{2}, z_{2})$$
  $$\vec{p}\cdot\vec{q}=(x_{1}x_{2}, y_{1}y_{2}, z_{1}z_{2})$$
  
*  Norma de un vector

$$\lVert p \rVert = \sqrt{\vec{p}\cdot\vec{p}}$$

* Ángulo entre 2 vectores:

$$\vec{p} \cdot \vec{q}=\lVert p \rVert \lVert q \rVert \cos\theta $$
$$\cos\theta =  \frac{\vec{p} \cdot \vec{q}}{\lVert p \rVert \lVert q \rVert}$$
